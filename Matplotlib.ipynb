```python
import matplotlib.pyplot as plt

# 1. Línea simple
plt.plot([7,7.5,8], [9,40,60])
plt.title("Línea simple")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(['12 años', '13 años', '14 años'], [5, 8, 22])
plt.title("Edades 7to primaria")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.pie([5, 8, 22], labels=['12 años', '13 años', '14 años'])
plt.title("Edades 7to primaria")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.barh(['Metabolito A', 'Metabolito B', 'Metabolito C'], [0.05, 0.07, 0.03])
plt.title("Cantidad metabolito en porcentaje")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.pie([0.05, 0.07, 0.03], labels=['Metabolito A', 'Metabolito B', 'Metabolito C'])
plt.title("Cantidad metabolito en porcentaje")
plt.show()
```
