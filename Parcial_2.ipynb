```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")


np.random.seed(42)
ninos = np.random.normal(loc=130, scale=10, size=30)
adultos = np.random.normal(loc=170, scale=8, size=30)
adultos_mayores = np.random.normal(loc=165, scale=9, size=30)


df = pd.DataFrame({
    'Altura': np.concatenate([ninos, adultos, adultos_mayores]),
    'Grupo': ['NiÃ±os']*30 + ['Adultos']*30 + ['Adultos Mayores']*30
})


plt.figure(figsize=(8, 6))
sns.boxplot(x='Grupo', y='Altura', data=df, palette='Set2', showfliers=True)


plt.title('ComparaciÃ³n de alturas por grupo', fontsize=16)
plt.xlabel('Grupo', fontsize=12)
plt.ylabel('Altura (cm)', fontsize=12)


plt.show()
```

Pregunta #5

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
meses = pd.date_range("2024-01-01", periods=12, freq="M").strftime('%b')
regiones = ["Norte", "Centro", "Sur"]
datos = []
for mes in meses:
    for region in regiones:
        ventas = np.random.randint(1000, 5000)
        datos.append([mes, region, ventas])

df = pd.DataFrame(datos, columns=["Mes", "Region", "Ventas"])
ventas_totales = df.groupby("Mes", as_index=False)["Ventas"].sum()
ventas_totales["Promedio_movil"] = ventas_totales["Ventas"].rolling(window=3).mean()
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.barplot(data=df, x="Mes", y="Ventas", hue="Region", palette="Set2", ci=None)
plt.plot(
    ventas_totales["Mes"],
    ventas_totales["Promedio_movil"],
    marker="o",
    color="black",
    linewidth=2,
    label="Promedio mÃ³vil"
)
plt.ylabel("Ventas")
plt.xlabel("Mes")
plt.title("Ventas mensuales por regiÃ³n con promedio mÃ³vil")
plt.legend(title="", labels=["Ventas totales", "Promedio mÃ³vil"], loc="upper left")
plt.tight_layout()
plt.show()
```
